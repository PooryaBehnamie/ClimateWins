```python
# Check for duplicats
df_dup = df.loc[df.duplicated()]
df_dup
```

```python
# Check for columns with inconsistent data types in specific columns (chatGPT version)
for col in df.columns.tolist():
    
    # Check if there are any inconsistencies in data types within the column
    weird = (df[[col]].apply(lambda x: x.map(type)) != df[[col]].iloc[0].apply(type)).any(axis=1)
    
    # If inconsistencies are found, print the column name
    if len(df[weird]) > 0:
        print(col)
    else:
        print(f'There is no mixed-type data in {col}')
```

### 2. Scaling

```python
#create a scaler object using StandardScaler from sklearn.preprocessing
#StandardScaler assumes data is normally distributed and scales with a distribution
#around 0 and a standard deviation of 1. Scaling happens independently with each variable.

scaler = StandardScaler()
```

```python
#create new df with scaled data

df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled.head()
```

```python
df_scaled.columns = df.columns
df_scaled.head(10)
```
